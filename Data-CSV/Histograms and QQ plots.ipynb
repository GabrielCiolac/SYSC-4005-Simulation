```python
values_to_be_inversed = []
for x in range(1,301) :
    values_to_be_inversed.append((x-0.5)/300)
    
```

```python
inversed_values = stats.lognorm(1, scale=np.exp(1.5)).ppf(values_to_be_inversed)
plt.scatter(inversed_values,data[0], c='r')
plt.axline([0, 0], [1, 1])
plt.xlabel('Inverse Lognormal values')
plt.ylabel('Ordered values')
plt.title('servinsp1 QQ plot with lognormal  = 1 , ^2 = 1.5')
```

Formula to calculate bin width for each histogram

```python
## calculating bin width for servinsp1
Q1 = np.quantile(data[0], 0.25)
Q3 = np.quantile(data[0], 0.75)
IQR = Q3 - Q1
cube = np.cbrt(len(data[0]))
bin_width = (IQR/cube * 2)
print(bin_width)
```

Formula to count number of occurances in each bin

```python
servinsp1_array
bin_values = 0
bin_width_start = 0
bin_width_end = bin_width
for i in range (0,26):
    for data_value in servinsp1_array:
        if(data_value <= bin_width_end and data_value > bin_width_start):
            bin_values = bin_values + 1
    print('bin {} from {} to {} is {} '.format(i,bin_width_start,bin_width_end,bin_values))
    bin_values = 0
    bin_width_start = bin_width_end
    bin_width_end = bin_width_end + bin_width
```
